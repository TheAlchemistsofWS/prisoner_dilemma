```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_prisoner = pd.read_excel("dp_prisoner_2024-12-06.xlsx")
df_prisoner.head()
```

```python
# Удалим все строчки из датасета, у которых значение в столбце "participant.label" равно " ". Это первые 7 строчек
df_prisoner = df_prisoner.iloc[8:]
df_prisoner.head(10)
```

Задача первой игры была выявить обычные стратегии игроков. Давайте разобьем людей на группы, в зависимости от того, какую стратегию они использовали. Для нас наибольшую ценность несет именно первый ход игрока, так как зачастую следующие ходы игрока зависят от типа его соперника (какую стратегию он использует).


```python
df_prisoner_first_round = df_prisoner.iloc[:12]
df_prisoner_first_round
```

```python
cooperate_players = []
defect_players = []

# Распределяем игроков по спискам
for _, row in df_prisoner_first_round.iterrows():
    if row['player.cooperate'] == 1.0:
        cooperate_players.append(row['participant.label'])
    else:
        defect_players.append(row['participant.label'])

print("Игроки, выбравшие кооперацию:")
for player in cooperate_players:
    print(f"- {player}")

print("\nИгроки, выбравшие оппортунизм:")
for player in defect_players:
    print(f"- {player}")
```

Посмотрим, как изменились стратегии игроков во втором раунде.

```python
cooperate_players_2 = []
defect_players_2 = []

# Распределяем игроков по спискам
for _, row in df_prisoner_second_round.iterrows():
    if row['player.cooperate'] == 1.0:
        cooperate_players_2.append(row['participant.label'])
    else:
        defect_players_2.append(row['participant.label'])

print("Игроки, выбравшие кооперацию:")
for player in cooperate_players_2:
    print(f"- {player}")

print("\nИгроки, выбравшие оппортунизм:")
for player in defect_players_2:
    print(f"- {player}")
```

Посчитаем, сколько кооперативных ходов сделали игроки за всю игру

```python
cooperative_moves = {}

for _, row in df_prisoner.iterrows():
    player_name = row['participant.label']
    is_cooperative = row['player.cooperate']

    # Пропускаем строки без имени игрока
    if pd.isna(player_name):
        continue

    # Если игрок еще не в словаре - добавляем его
    if player_name not in cooperative_moves:
        cooperative_moves[player_name] = 0

    # Если ход был кооперативным - увеличиваем счетчик
    if is_cooperative == 1.0:
        cooperative_moves[player_name] += 1

# Сортируем словарь по именам игроков
sorted_moves = dict(sorted(cooperative_moves.items()))

# Выводим результаты
print("Количество кооперативных ходов по игрокам:")
for player, moves in sorted_moves.items():
    print(f"- {player}: {moves} кооперативных ходов")
```

Давайте посчитаем процент кооперативных ходов в каждом раунде.

```python
# Создаем словарь для хранения статистики по раундам
round_stats = {}

# Проходим по всем строкам датафрейма
for _, row in df_prisoner_tax.iterrows():
    round_num = row['subsession.round_number']
    is_cooperative = row['player.cooperate']

    # Если раунд еще не в словаре - добавляем его
    if round_num not in round_stats:
        round_stats[round_num] = {'cooperative': 0, 'total': 0}

    # Увеличиваем счетчики
    round_stats[round_num]['total'] += 1
    if is_cooperative == 1.0:
        round_stats[round_num]['cooperative'] += 1

# Выводим результаты
print("Процент кооперативных ходов по раундам:")
for round_num in sorted(round_stats.keys()):
    stats = round_stats[round_num]
    cooperative_percentage = (stats['cooperative'] / stats['total']) * 100
    print(f"Раунд {round_num}: {cooperative_percentage:.1f}% кооперативных ходов")
```

В последних четырех раундах процент кооперативных ходов снизился. Это можно объяснить тем, что в них мы перестали выплачивать субсидию игрокам, то есть мы как бы перестали страховать игроков в случае если их захотят обмануть.
